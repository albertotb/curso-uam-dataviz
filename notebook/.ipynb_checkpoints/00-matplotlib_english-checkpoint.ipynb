# Introduction to Matplotlib

The `matplotlib` library has classes and functions to create plots with different styles and formats:

Reference: http://matplotlib.org/   
FAQ : https://matplotlib.org/faq/usage_faq.html    

The most basic function is [`plt.plot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html). This will plot a line chart between two sequences:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.Series(range(-10, 11))
y = x ** 2

print(x)
```

```python
plt.plot(x, y, color="red")
plt.title("x**2 function")
plt.xlabel("x")
plt.ylabel("x**2")
plt.show()
```

```python
plt.plot(x, y, color="red", label="x**2")
plt.plot(x, x ** 3, color="blue", label="x**3")
plt.title("Square vs cubic")
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color="red", label="x**2")
plt.plot(x, x ** 3, color="blue", label="x**3")
plt.title("Square vs cubic")
plt.xlabel("x")
plt.xlim([-10, 10])
plt.ylim([-100, 100])
plt.legend()
plt.show()
```

## Setting colors, linewidths, linetypes and markers

With matplotlib, we can define the colors of lines, type of line and width using arguments of the plot function. Lists of available line colors and styles:

  - [Line colors](https://matplotlib.org/stable/gallery/color/named_colors.html)
  - [Line styles](https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html)


```python
plt.plot(x, y, color="tab:purple", linestyle="-.")
plt.show()
```

We can also change the default width of the line with the `linewidth` parameter. It takes a float: 1 is the default line width, 2 is twice the default line width and 10 is ten times the default line width (it can also be less than 1 to reduce the default line width)

```python
plt.plot(x, y, color="tab:purple", linestyle="--", linewidth=4)
plt.show()
```

When plotting data as lines, we are just plotting points and matplotlib is filling up the rest. If we want to enphasize that, we can add markers for the underlying points. There are [several styles](https://matplotlib.org/stable/api/markers_api.html) that we can use:

```python
plt.plot(x, y, color="tab:purple", marker="o")
plt.show()
```

## Saving figures

Plots can be saved using the [`plt.savefig()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html) function. The current plot will be saved. Note that you call this function **instead of** `plt.show()`, but you **can not** call both

```python
plt.plot(x, y, color="red", label="x**2")
plt.plot(x, x ** 3, color="blue", label="x**3", linestyle="-.")
plt.title("Square vs cubic")
plt.xlabel("x")
plt.xlim([-10, 10])
plt.ylim([-100, 100])
plt.legend()
plt.savefig("square_cubic.pdf")
```

## Types of plots

There are several types of plots that can be done using `matplotlib`. The main ones are:

  - Histogram: [`plt.hist()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)
  - Barplot: [`plt.bar()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
  - Pie chart: [`plt.pie()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)
  - Scatterplot: [`plt.scatter()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)
  - Boxplot: [`plt.boxplot()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

### Histograms

Histograms are used to display distributions of continuous numeric variables:

```python
import random

n = 100000
normal = []
for i in range(n):
    normal.append(random.gauss(5, 3))

plt.hist(normal, bins=50, color="blue", range=(-5, 15))
plt.title("Gaussian distribution")
plt.show()
```

A lot of work is being done underneath to create a histogram:

  - Sort the original values
  - Decide how many intervals (bars) to use
  - Count how many values fall into each interval (bar)
  - Plot count of values as height of each bar
  
It is very important to always test different values of the `bins` parameter, that represents the number of intervals (bars) to use

### Bar plots

Barplots are used to display the relationship between a numeric and discrete variable:

```python
x_values = [5, 6, 3, 7, 2]
y_values = ["A", "B", "C", "D", "E"]
plt.bar(y_values, x_values, color = "green")
plt.show()
```

We can also use horizontal bars by changing the function to `plt.barh()`:

```python
plt.barh(y_values, x_values, color = "green")
plt.show()
```

### Pie charts

Pie charts focus on showing parts of a whole. The entire pie would be 100 percent. The question is how much of that percentage each value occupies. The following example shows how to create a pie chart with many of the special features in place:

```python
values = [5, 8, 9, 10, 4, 7]
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = ['A', 'B', 'C', 'D', 'E', 'F']
explode = (0, 0.1, 0, 0, 0, 0)

plt.pie(values, colors=colors, labels=labels, explode=explode, counterclock=False)
plt.title('Pie chart example')
plt.show()
```

### Scatterplots

Scatterplots represent two numerical variables against each other. They can be useful to show clusters of data.

```python
n = 100
numbers = []
mu_list = []
for i in range(n):
    mu = random.choice([0, 5])
    numbers.append(random.gauss(mu, 1))
    mu_list.append(mu)
```

```python
plt.scatter(range(n), numbers)
plt.show()
```

A very useful technique is to map each point to a different color. In this example we will color points differently if they were generated with the `mu=5` or `mu=0` distributions:

```python
plt.scatter(range(n), numbers, c=mu_list)
plt.show()
```

The purpose of a scatterplot is to help you see patterns in data. Besides lists, we can also pass directly pandas Series. Let us see an example with real data:

```python
df = pd.read_csv("../data/california_housing_train.csv")
```

```python
df.columns
```

```python
plt.scatter(df["median_house_value"], df["median_income"])
plt.xlabel("Median house value")
plt.ylabel("Median income")
plt.show()
```

There are many points (households) in the previous graph. Two useful parameters that sometimes help when dealing with many points are:

   - `alpha`: changes the transparency of the points, number between 0 (100% transparent) and 1 (100% opaque)
   - `s`: changes the sizes of the points

```python
plt.scatter(df["longitude"], df["latitude"], alpha=0.5, s=10)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

Again, we can color the points by another variable, for instance `median_house_value`:

```python
plt.scatter(df["longitude"], df["latitude"], alpha=0.5, s=10, c=df["median_house_value"])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar()
plt.show()
```

### Boxplots

Boxplots represent the main statistics of a numerical variable:

<img src="https://www.simplypsychology.org/boxplot.jpg" alt="boxplot" width="700"/>

```python
plt.boxplot(df["median_house_value"], vert=False)
plt.show()
```

The black points outside the "whiskers" of the plot are extreme values (they are omitted from the computation of the minimum and maximum). The `vert` argument determines if the boxplot is horizontal or vertical.

## Figure vs axes

<img src="https://www.earthdatascience.org/images/earth-analytics/plot-data/fig-2-plots.png" alt="fig_vs_axes" width="700"/>

While most matplotib pyplot functions translate directly to axes methods (such as `plt.plot()` → `ax.plot()`, `plt.legend()` → `ax.legend()`, etc.), this is not the case for all commands. In particular, functions to set limits, labels, and titles are slightly modified. For transitioning between them, make the following changes:

 - `plt.xlabel()` → `ax.set_xlabel()`
 - `plt.ylabel()` → `ax.set_ylabel()`
 - `plt.xlim()` → `ax.set_xlim()`
 - `plt.ylim()` → `ax.set_ylim()`
 - `plt.title()` → `ax.set_title()`
